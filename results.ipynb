Construindo o gr√°fico de barras para os resultados da CNN para os motores 1, 2 e 3

```python
metrics = cnn_m1.columns
r1 = np.arange(len(metrics))
r2 = [x + 0.25 for x in r1]
r3 = [x + 0.25 for x in r2]

plt.figure(figsize=(10,10))
plt.bar(r1, cnn_m1.values[0], color='#6A5ACD', width=0.25, label='Engine 1')
plt.bar(r2, cnn_m2.values[0], color='#6A95ED', width=0.25, label='Engine 2')
plt.bar(r3, cnn_m3.values[0], color='#00BFFF', width=0.25, label='Engine 3')
plt.xlabel("Metrics")
plt.ylabel("Values")
plt.title("Value of Metrics for Engine 1, Engine 2, and Engine 3")
plt.legend()
plt.show()
```
